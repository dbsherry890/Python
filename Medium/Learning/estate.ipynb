## Sets

```python
a = {1,2,3,4,5}
b= {3,4,5,6,7}
```

```python
a.union(b) # {1, 2, 3, 4, 5, 6, 7}  same as a | b
a.intersection(b) # {3, 4, 5} same as a & b
a.difference(b) # {1, 2} same as a - b (elements in a that are not in b)
a.symmetric_difference(b) # {1, 2, 6, 7} same as a ^ b (all elements that they don't share)
a.issubset(b) # False. Checking if all elements of a are in b
a.issuperset(b) # False. Checking if all elements in b are in a
a.isdisjoint(b) # False. Checking if a and b have no elements in common
a == b # checks if sets contain the same elements. Order doesn't matter
```

```python
a.remove(6) # raises error. Use discard() to suppress error
a.pop() # removes random element
a.clear() # removes all elements
```
