```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns

np.set_printoptions(suppress=True, precision = 2)

nba = pd.read_csv('./nba_games.csv')

# Subset Data to 2010 Season, 2014 Season
nba_2010 = nba[nba.year_id == 2010]
nba_2014 = nba[nba.year_id == 2014]
```

```python
nba_2010.head()
```

```python
knicks_pts_10 = nba_2010.pts[nba_2010.fran_id == 'Knicks']
nets_pts_10 = nba_2010.pts[nba_2010.fran_id == 'Nets']
knicks_pts_10
# nets_pts_10
```

```python
knicks_pts_14 = nba_2014.pts[nba_2014.fran_id == 'Knicks']
nets_pts_14 = nba_2014.pts[nba_2014.fran_id == 'Nets']
knicks_pts_14
# nets_pts_10
```

```python
diff_means_2010 = np.mean(knicks_pts_10) - np.mean(nets_pts_10)
diff_means_2010
```

```python
diff_means_2014 = np.mean(knicks_pts_14) - np.mean(nets_pts_14)
diff_means_2014
```

```python
# Overlapping histograms
plt.hist(knicks_pts_10, alpha=0.5, label='Knicks')
plt.hist(nets_pts_10, alpha=0.5, label='Nets')
plt.title('Overlapping')
plt.show()
```

```python
# Overlapping histograms
plt.hist(knicks_pts_14, alpha=0.5, label='Knicks', density=True)
plt.hist(nets_pts_14, alpha=0.5, label='Nets', density=True)
plt.title('Overlapping')
plt.show()
```
